```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
df = pd.read_csv("/content/drive/MyDrive/My_Project/ds_salaries.csv")  # adjust file name if different

# Step 3: Filter only analyst roles
df = df[df['job_title'].str.contains("analyst", case=False)]

# Step 4: Basic EDA
print(df.info())
print(df['experience_level'].value_counts())

# Step 5: Salary by Experience Level
sns.barplot(x='experience_level', y='salary_in_usd', data=df)
plt.title("Average Salary by Experience Level")
plt.show()

# Step 6: Save cleaned version for dashboard
df.to_csv("/content/drive/MyDrive/My_Project/analyst_data.csv", index=False)
```
